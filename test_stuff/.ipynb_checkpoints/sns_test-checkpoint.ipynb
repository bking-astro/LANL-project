```python
EOS_small = pd.read_table('../data/EOSCEFTVE1.dat', header=None).to_numpy()
```

```python
nsim_1 = 10
size_MRL = 80
nsamp = 5
size_ex = 5*EOS_small.shape[0]
size_1 = size_ex + EOS_small.shape[0] -1

EOS_snstest = np.zeros((size_1, 3, nsim_1))
causality = np.zeros(nsim_1, dtype=bool)
MRL_snstest = np.zeros((size_MRL, 3, nsim_1))

for i in range(nsim_1):
    EOS_snstest[:,:,i], causality[i]  =  EOS_extension_test.extend(EOS_small, nsamp, size=size_ex, ext_type='polytrop')
    MRL_snstest[:,:,i] = test_TOVsolver.solve(EOS_snstest[:,:,i], size=size_MRL)
```

```python
nrow = nsim_1*size_MRL
ncol = 4

test_nparray = np.zeros((nrow, ncol), dtype=float)

k=0
for i in range(nsim_1):
    for j in range(size_MRL):
        test_nparray[k,0] = float(i)
        test_nparray[k,1] = MRL_snstest[j,0,i]
        test_nparray[k,2] = MRL_snstest[j,1,i]
        test_nparray[k,3] = MRL_snstest[j,2,i]
        k +=1
```

```python
test_df = pd.DataFrame(data=test_nparray, columns=['nsim', 'mass', 'radius', 'lambda'])
test_df.head()
```

```python
sns.relplot(data=test_df, x="mass", y="radius", hue='nsim', kind='line')
```

```python
# load in the big simulation we did

M_1k = np.loadtxt("../data/MRL1ksim_cs_M.dat")
R_1k = np.loadtxt("../data/MRL1ksim_cs_R.dat")
L_1k = np.loadtxt("../data/MRL1ksim_cs_L.dat")

nsim = len(M_1k[0,:])
size_MRL = len(M_1k[:,0])

nrow = nsim*size_MRL
ncol = 4

test1k_nparray = np.zeros((nrow, ncol), dtype=float)

k=0
for i in range(nsim):
    for j in range(size_MRL):
        test1k_nparray[k,0] = i
        test1k_nparray[k,1] = M_1k[j,i]
        test1k_nparray[k,2] = R_1k[j,i]
        test1k_nparray[k,3] = L_1k[j,i]
        k +=1
```

```python
test1k_df = pd.DataFrame(data=test1k_nparray, columns=['nsim', 'mass', 'radius', 'lambda'])
test1k_df.head()
```
