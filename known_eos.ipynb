```python
plt.plot(SLy_MR[:,0], SLy_MR[:,1], label="SLy")
plt.plot(H4_MR[:,0], H4_MR[:,1], label="H4")
plt.plot(MS1_MR[:,0], MS1_MR[:,1], label="MS1")
plt.plot(MPA1_MR[:,0], MPA1_MR[:,1], label="MPA1")

plt.plot(cs_eq1MRL[:,1], cs_eq1MRL[:,0], label='cs=1', color='black')
plt.xlabel("Radius (km)")
plt.ylabel("Mass (Mo)")
plt.legend()
```

```python
SLy_ML = np.loadtxt("/Users/brendanking/Documents/LANL/LANL_project/data/MLam_SLy.dat")
H4_ML = np.loadtxt("/Users/brendanking/Documents/LANL/LANL_project/data/MLam_H4.dat")
MS1_ML = np.loadtxt("/Users/brendanking/Documents/LANL/LANL_project/data/MLam_MS1.dat")
MPA1_ML = np.loadtxt("/Users/brendanking/Documents/LANL/LANL_project/data/MLam_MPA1.dat")
```

```python
plt.plot(SLy_ML[:,0], SLy_ML[:,1], label="SLy")
plt.plot(H4_ML[:,0], H4_ML[:,1], label="H4")
plt.plot(MS1_ML[:,0], MS1_ML[:,1], label="MS1")
plt.plot(MPA1_ML[:,0], MPA1_ML[:,1], label="MPA1")

ax = plt.gca()
plt.vlines(1.4, ymin=ax.get_ylim()[0], ymax=ax.get_ylim()[1], ls='dashed', color='black')

plt.ylabel(r"$\tilde{\Lambda}$ (unitless)")
plt.yscale("log")
plt.xlabel("Mass (Mo)")
plt.legend()
```
